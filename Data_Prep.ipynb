```python
import librosa 
import librosa.display 
import numpy as np 
import matplotlib.pyplot as plt
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
file = "blues.00001.wav"
```

# Waveform

```python
señal,sr= librosa.load(file,sr=22050)
librosa.display.waveshow(señal,sr=sr)
plt.xlabel("tiempo")
plt.ylabel("Amplitud")
plt.show()
```

# FFT Spectogram 1/2

```python
fft = np.fft.fft(señal)
Magnitud = np.abs(fft)
Frecuencia = np.linspace(0, sr, len(Magnitud))
Frecuencia_Izq = Frecuencia[:int(len(Frecuencia)/2)]
Magnitud_Izq= Magnitud[:int(len(Magnitud)/2)]

plt.plot (Frecuencia_Izq, Magnitud_Izq) 
plt.xlabel("Frecuencia")
plt.ylabel ("Magnitud") 
plt.show()
```

# SFFT Spectrogram 

```python
n_fft= 2048
hop_length = 512
stft = librosa.core.stft(señal, hop_length=hop_length, n_fft=n_fft)
spectrogram = np.abs (stft)
log_spectrogram=librosa.amplitude_to_db(spectrogram)

librosa.display.specshow(log_spectrogram, sr=sr, hop_length=hop_length)

plt.xlabel("tiempo") 
plt.ylabel("Frecuencia") 
plt.colorbar() 
plt.show()
```

```python
n_fft = 2048
hop_length = 512
n_mfcc = 13
MFFCs = librosa.feature.mfcc(y=señal, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mfcc=n_mfcc)

# Visualizar MFFCs
librosa.display.specshow(MFFCs, sr=sr, hop_length=hop_length)
plt.ylabel('MFFC')
plt.xlabel('Time')
plt.show()
```
